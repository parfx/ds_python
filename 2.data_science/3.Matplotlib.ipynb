- простые линейные графики

```python
# массив (вектор-строка) из 1000 значений
# линейная последовательность между 0 и 10
x = np.linspace(0, 10, 1000)
x
```

```python
# нарисуем график функции
# например, график функции синуса
y = np.sin(x)
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
# совмещение двух графиков в одном окне
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
# Настройка графиков: цвета и вид линий

# цвета можно задавать разными способами
plt.plot(x, np.cos(x), color='blue') # по названию
plt.plot(x, np.cos(x - 1), color='g') # первая буква цвета (rgbcmyk)
plt.plot(x, np.cos(x - 2), color='0.75') # оттенки серого в диапазоне от 0 до 1
plt.plot(x, np.cos(x - 3), color='#ffdd44') # 16-ричный код (RGB от 00 до FF)
plt.plot(x, np.cos(x - 4), color=(1.0, 0.2, 0.3)) # RGB-кортеж, значений от 0 до 1
plt.plot(x, np.cos(x - 5), color='chartreuse') # название цвета HTML
plt.show()
```

```python
# стиль линий
plt.plot(x, x + 5, linestyle='-') #'solid') # сплошная
plt.plot(x, x + 6, linestyle='--') #'dashed') # штриховая
plt.plot(x, x + 7, linestyle=':') #'dotted') # пунктирная
plt.plot(x, x + 8, linestyle='-.') #'dashdot') # штрих-пунктирная
```

```python
# объединение кодов стиля и цвета линий в одном аргументе
plt.plot(x, np.sin(x), '-g')
plt.plot(x, np.sin(x + 1), '--r')
plt.plot(x, np.sin(x + 2), ':k')
plt.show()
```

```python
# пределы осей
plt.plot(x, np.sin(x), '--g')

# индивидуальная настройка осей 
plt.ylim(-2, 10)
plt.xlim(0, 20)
plt.show()
```

```python
# настройка обеих осей
plt.plot(x, np.sin(x), '--g')
plt.axis([0, 5, -1.5, 1.5])
```

```python
# метки на графиках

# интерактивный режим
%matplotlib notebook

# название (заголовок) графика и метки его осей
plt.plot(x, np.sin(x), '-')

plt.title("Синусоида")
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
# легенда графика

plt.plot(x, np.sin(x), '-r', label='sin')
plt.plot(x, np.cos(x), '-g', label='cos')

plt.legend()
plt.title("Синусоида")
plt.xlabel('x')
plt.ylabel('sin(x)')

plt.show()
```

- Гистограммы

```python
# data = np.random.random(1000)
data = np.random.randn(1000)

plt.hist(data, bins=50)
plt.show()
```

```python
# настройки
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

plt.hist(x1, bins=50, alpha=0.5, color='r')
plt.hist(x2, bins=50, alpha=0.3, color='g', edgecolor='k')
plt.hist(x3, bins=50, alpha=0.5)

plt.show()
```

```python
x = np.random.randn(50)
y = np.random.randn(50)

sizes = np.random.randn(50) * 100


plt.scatter(x, y, marker='o', s=sizes)
plt.show()
```

```python
x = np.random.randn(50)
y = np.random.randn(50)

sizes = np.random.randn(50) * 100
colors = np.random.randn(50)


plt.scatter(x, y, marker='o', s=sizes, c=colors, cmap='viridis')
plt.show()
```

```python
plt.scatter(x, y, marker='o', s=sizes, c=colors, cmap='viridis', alpha=0.5)
plt.show()
```
